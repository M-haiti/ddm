```python
"""
Librer√≠as de rutinas importadas
"""
import imageio
import matplotlib.pyplot as plt
import numpy as np

#Lectura de archivo imagen
pic = imageio.imread('images/test_cut.jpeg')

#Creaci√≥n de figura Plot
plt.figure(figsize = (100,100))

#Mostrar imagen dentro de plot
plt.imshow(pic)

print('Tipo de imagen : ' , type(pic)) 
print('Forma de imagen, largo, ancho y capas RGB : {}'.format(pic.shape)) 
print('Imagen - Alto {}'.format(pic.shape[0])) 
print('Imagen - Ancho {}'.format(pic.shape[1])) 
print('Capas de imagen {}'.format(pic.ndim))
print('Tama√±o de imagen {}'.format(pic.size)) 
print('Maximo valor RGB {}'.format(pic.max())) 
print('Minimo valor RGB {}'.format(pic.min()))
```

```python
#Grayscale
gray = lambda rgb : np.dot(rgb[... , :3] , [0.299 , 0.587, 0.114]) 
gray = gray(pic)  
plt.figure( figsize = (10,10))
plt.imshow(gray, cmap = plt.get_cmap(name = 'gray'))
plt.show()

limite = 170
fondo = 0
captura = 0
blanco = 0

for i in range(pic.shape[0]):
    for j in range(pic.shape[1]):
        #iterando por cada pixel, analizando el valor de gris
        if gray[i,j]<10:
            fondo += 1
            #colorizar pixel verde
            pic[ i , j ,1 ] = 255 
            
        elif gray[i,j]<limite:
            captura +=1
            #colorizar pixel rojo
            pic[ i , j ,0 ] = 255
            
        else:
            blanco+=1
            #colorizar pixel azul
            pic[ i , j ,2 ] = 255 

total = (gray.shape[0] * gray.shape[1])-fondo

porcentaje_blanco = (blanco/total)*100
porcentaje_blanco = "{:.2f}".format(porcentaje_blanco)

porcentaje_captura = (captura/total)*100
porcentaje_captura = "{:.2f}".format(porcentaje_captura)

print(f'Total de pixeles: {total}, 100%')
print(f'Pixeles encima de limite: {blanco}, Porcentaje: {porcentaje_blanco}%')
print(f'Pixeles captura: {captura}, Porcentaje: {porcentaje_captura}%')

plt.figure(figsize = (100,100))
plt.imshow(pic)
plt.savefig('analisis.png', bbox_inches='tight')           


    
```
