```python
import imageio
import matplotlib.pyplot as plt
import numpy as np

#%matplotlib inline
pic = imageio.imread('images/test_cut.jpeg')
plt.figure(figsize = (100,100))
plt.imshow(pic)

print('Type of the image : ' , type(pic)) 
print('Shape of the image : {}'.format(pic.shape)) 
print('Image Height {}'.format(pic.shape[0])) 
print('Image Width {}'.format(pic.shape[1])) 
print('Dimension of Image {}'.format(pic.ndim))
print('Image size {}'.format(pic.size)) 
print('Maximum RGB value in this image {}'.format(pic.max())) 
print('Minimum RGB value in this image {}'.format(pic.min()))
```

```python

#Grayscale
gray = lambda rgb : np.dot(rgb[... , :3] , [0.299 , 0.587, 0.114]) 
gray = gray(pic)  

plt.figure( figsize = (10,10))
plt.imshow(gray, cmap = plt.get_cmap(name = 'gray'))
plt.show()

limite = 170
fondo = 0
captura = 0
blanco = 0

for i in range(pic.shape[0]):
    for j in range(pic.shape[1]):
        if gray[i,j]<10:
            fondo += 1
            pic[ i , j ,1 ] = 225 
        elif gray[i,j]<limite:
            captura +=1
            pic[ i , j ,0 ] = 225 
        else:
            blanco+=1
            pic[ i , j ,2 ] = 225 

total = (gray.shape[0] * gray.shape[1])-fondo

porcentaje_blanco = (blanco/total)*100
porcentaje_blanco = "{:.2f}".format(porcentaje_blanco)

porcentaje_captura = (captura/total)*100
porcentaje_captura = "{:.2f}".format(porcentaje_captura)

print(f'Total de pixeles: {total}, 100%')
print(f'Pixeles encima de limite: {blanco}, Porcentaje: {porcentaje_blanco}%')
print(f'Pixeles captura: {captura}, Porcentaje: {porcentaje_captura}%')

plt.figure(figsize = (100,100))
plt.imshow(pic)
plt.savefig('analisis.png', bbox_inches='tight')           


    
```
